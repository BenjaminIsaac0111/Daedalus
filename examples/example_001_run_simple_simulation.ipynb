```python
from vivarium import InteractiveContext

import daedalus.static as Static
import daedalus.static_h as StaticH
import daedalus.assignment as Assignment
from daedalus.population import Population
from daedalus.mortality import Mortality
from daedalus.migration import Migration

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Set the filename_example:
# Note: this needs to be changed in the Population class as well
# XXX TODO: read the filename from the input
filename_example = '../data/Testfile_002.csv'
sim = InteractiveContext('../config/model_specification.yaml', setup=False)
sim.configuration.update({'population': {'population_size': len(pd.read_csv(filename_example))}})
sim.setup()
```

```python
sim.run()
```

```python
# Read output files (generated at each time step)
population = np.loadtxt("./population.txt")
mortality = np.loadtxt("./mortality.txt")
migration = np.loadtxt("./migration.txt")
```

```python
plt.figure(figsize=(10, 6))

plt.plot(population, 'k', lw=3, label="population")
plt.text(len(population), population[-1], int(population[-1]), size=16)

plt.plot(mortality, 'r', lw=3, label="mortality")
plt.text(len(mortality), mortality[-1], int(mortality[-1]), size=16, color='r')

plt.plot(migration, 'b', lw=3, label="migration")
plt.text(len(migration), migration[-1], int(migration[-1]), size=16, color='b')

plt.legend(prop={'size': 20})
plt.xlabel("Time (in days)", size=24)
plt.ylabel("Count", size=24)
plt.xticks(size=16)
plt.yticks(size=16)
plt.xlim(0, len(population)*1.1)
plt.grid()
```
